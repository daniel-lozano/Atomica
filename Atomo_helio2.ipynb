```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi
import math
```

```python
h=0.0001
l=0
z=2
Eo=(z**2)*(-13.6)
e=math.e
Umax=100000
U=np.zeros([Umax,1],dtype=float)
```

```python
R=np.zeros([Umax,1],dtype=float)
R[0]=1
Rprime=np.zeros([Umax,1],dtype=float)
Rprime[0]=(-1)
Rdprime=np.zeros([Umax,1],dtype=float)
Rdprime[0]=(0.0001)
Po=np.zeros([Umax,1],dtype=float)

for i in range(Umax):
    U[i]=((i+1)*h)

```

```python
def potencial(i):
    P=(-1-e**(-U[i]))/U[i]
    return -(1+e**(-U[i]))/U[i]
```

```python
"""
llenar el potencial
"""
for i in range(shape(U)[0]):
    Po[i]=potencial(i)

plot(U,Po,label="Potencial")
xlim(0,0.0100)
legend()
plt.show()
plt.close()

plot(U,Po,"ko",label="Potencial")
xlim(0,0.0020)
plt.show()
plt.close()
```

```python
E=25.43
```

```python
def func_prime_1(x, y_1, y_2,i):
    return y_2

def func_prime_2(x, y_1, y_2,i):
    return -(2.0/x)*y_2+y_1*(-E/Eo+Po[i])

def RungeKuttaSecondOrderStep(x_old, y_1_old, y_2_old,i):
    #get the first derivatives
    y_prime_1 = func_prime_1(x_old,y_1_old, y_2_old,i)
    y_prime_2 = func_prime_2(x_old,y_1_old, y_2_old,i)
    
    #from this estimation move to the middle point
    x_middle = x_old+ (h/2.0)
    y_1_middle = y_1_old + (h/2.0) * y_prime_1
    y_2_middle = y_2_old + (h/2.0) * y_prime_2 
    
    #compute the derivatives at the middle point
    y_middle_prime_1 = func_prime_1(x_middle, y_1_middle, y_2_middle,i)
    y_middle_prime_2 = func_prime_2(x_middle, y_1_middle, y_2_middle,i)
    
    x_new = x_old + h
    y_1_new = y_1_old + h * y_middle_prime_1 
    y_2_new= y_2_old + h * y_middle_prime_2
    return x_new, y_1_new, y_2_new
    
```

```python
U[0]=0.0001
R[0] = 1.0
Rprime[0]=-1.0    

for i in range(1,Umax):
    U[i],R[i],Rprime[i] = RungeKuttaSecondOrderStep(U[i-1], R[i-1],Rprime[i-1],i-1)
```

```python
plot(U,R,label="R(u)")
title("energia:"+str(E))
xlabel("u")
ylabel("R")
savefig("Radial1.pdf")
legend()
```

```python
proba=np.zeros([Umax,1],dtype=double)
sumatoria=np.zeros([Umax,1],dtype=double)
probaNor=np.zeros([Umax,1],dtype=double)
norma=np.zeros([Umax,1],dtype=double)
integral=np.zeros([Umax,1],dtype=double)
```

```python
for i in range(Umax):
    proba[i]=4*math.pi*(U[i]*R[i])**2

suma=proba[0]    

for i in range(1,Umax):
    suma=proba[i]*h+suma
    
for i in range(Umax):
   norma[i]=R[i]/((suma)**0.5)

for i in range(Umax):
    
    probaNor[i]=4*math.pi*(U[i]*norma[i])**2
```

```python
plot(U,proba,label="probabilidad")
legend()
savefig("probabilidad1.pdf")
plt.show()
plt.close()
plot(U,probaNor,label="probabilidad normalizada")
legend()
savefig("probabilidad1nor.pdf")
```

```python
Po[0]=0
V=np.zeros([Umax,1],dtype=double)
for i in range(1,Umax):
    Po[i]=+Po[i-1]+( (2-integral[i])*h*(U[i])**-2 )
V=Po-abs(Po[-1]+1/U[-1])
```
